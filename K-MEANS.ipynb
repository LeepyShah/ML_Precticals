<h1>Write a Python script that implements the Elbow Method to determine the optimal number of clusters for
a generated dataset. (Generate dataset of your choice with 14 points)
</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate dataset (14 points)
np.random.seed(42)
data = np.random.randint(10, 100, size=(14, 2))

# Elbow Method
wcss = []
for k in range(1, 8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)

# Plot
plt.figure()
plt.plot(range(1, 8), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

<h1>Use make_blobs to generate 300 samples. Analyse the data and apply KMeans clustering and generate the
graph for better representation. Show what happens if you change n_clusters count from 2 to
7(2,3,4,5,6,7)?
</h1>

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


#Generate Data
X,y=make_blobs(n_samples=300,random_state=42)
```

```python
#Try cluster counts from 2 to 7
plt.figure(figsize=(15,8))
for k in range(2,8):
    kmeans=KMeans(n_clusters=k,random_state=42)
    labels=kmeans.fit_predict(X)
    plt.subplot(2,3,k-1)
    plt.scatter(X[:,0], X[:,1], c=labels)
    plt.title(f"K = {k}")
plt.tight_layout()
plt.show()    
```

<h1>A grocery store wants to analyze customer purchase patterns to improve product placements. Generate
random transaction data (products purchased together), apply K-means clustering to group similar
transactions.</h1>

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
#Randome transactions data
np.random.seed(42)
data=pd.DataFrame({
    "milk":np.random.randint(0,2,100),
    "bread":np.random.randint(0,2,100),
    "butter":np.random.randint(0,2,100),
    "fruits":np.random.randint(0,2,100)    
})
#Apply Kmeans
kmeans=KMeans(n_clusters=3,random_state=42)
data['cluster']=kmeans.fit_predict(data)
# Plot using 2 features for visualization
plt.scatter(data["milk"], data["bread"], c=data["cluster"])
plt.xlabel("Milk")
plt.ylabel("Bread")
plt.title("Customer Transaction Clusters")
plt.show()

# Cluster summary
print(data.groupby("cluster").mean())
```

<h1>Generate random dataset with 35 points. Apply KMeans clustering and generate the plot with each cluster
points with different colours. What is the impact of varying cluster count? Justify your understanding with
suitable graphs.</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

np.random.seed(42)

data = pd.DataFrame({
    "Total_Spending": np.random.randint(100, 10000, 150),
    "Frequency": np.random.randint(1, 50, 150),
    "Avg_Basket_Size": np.random.randint(1, 20, 150)
})

kmeans = KMeans(n_clusters=3, random_state=42)
data["Cluster"] = kmeans.fit_predict(data)

plt.figure()
plt.scatter(data["Total_Spending"], data["Frequency"], c=data["Cluster"])
plt.xlabel("Total Spending")
plt.ylabel("Frequency")
plt.title("E-commerce Customer Clusters")
plt.show()

print(data.groupby("Cluster").mean())
```
